```python
# Word based tokenization
# https://www.nltk.org/api/nltk.tokenize.html
```

```python
import io
import nltk

try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download(
        'punkt',
        quiet=True,
        halt_on_error=False,
        raise_on_error=False,
        print_error_to=io.StringIO(),
    )

from nltk.tokenize import word_tokenize
```

```python
tokenization_input = "Hello, how are you doing today?"
tokens = word_tokenize(tokenization_input)

print(tokens)
```

```python
# Character based tokenization
character_tokens = list(tokenization_input)
print(character_tokens)
```

```python
# Sub word based tokenization
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")

tokens = tokenizer.tokenize(tokenization_input)
print(tokens)
```

```python
# WordPiece model tokenization (BERT-style tokenizer)
from transformers import AutoTokenizer

model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

text = "The weather is beautiful today!"
tokens = tokenizer.tokenize(text)
inputs = tokenizer(text)

print("Tokens:", tokens)
print("Token IDs:", inputs["input_ids"])
print("Decoded:", tokenizer.decode(inputs["input_ids"]))
```

```python
# Here the above output 101 and 102 are special tokens for [CLS] and [SEP] respectively. The rest of the tokens are the tokenized words from the input text.
```
